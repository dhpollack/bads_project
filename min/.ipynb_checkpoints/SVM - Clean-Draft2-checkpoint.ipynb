```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, make_scorer, roc_auc_score
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV, train_test_split
from sklearn.externals import joblib
import time
from sklearn import svm
import time
from mpl_toolkits.mplot3d import Axes3D

from badsproject import BADS

%matplotlib inline

```

```python
def evaluate_SVMs(training_size=10000, validation_size=-1, gammas=np.array([0.003]),Cs=np.array([10,20,60,120]),
                  num_runs=1,seed=1234,print_progress=False,
                  plot_heatmap=False, thresholds=np.linspace(-1,1,20), show_T=False, print_conf=False):
    Cs = np.array(Cs)
    gammas = np.array(gammas)
    thresholds=np.array(thresholds)
    # So first let's get the data
    dummy = BADS() # Instead of modifying the main model further we use the dummy variant to access the functions
    dummy.create_datasets()
    X_train_big = dummy.X_train
    y_train_big = dummy.y_train
    X_train_cv = None#dummy.X_train_cv
    y_train_cv = None #dummy.y_train_cv
        
    def split(X,y,sample_size):
        # This will create a balanced split. 
        A = X[y==1]
        B = X[y!=1]
        Ay = y[y==1]
        By = y[y!=1]
        tmp1 = np.random.permutation(A.shape[0])
        tmp2 = np.random.permutation(B.shape[0])
        athresh=int(sample_size*(0.2))
        bthresh=int(sample_size*(0.8))
        Xtrain = np.concatenate([A[tmp1,:][:athresh],B[tmp2,:][:bthresh]])
        ytrain = np.concatenate([Ay[tmp1][:athresh],By[tmp2][:bthresh]])
        Xval = np.concatenate([A[tmp1,:][athresh:(2*athresh)],B[tmp2,:][bthresh:(2*bthresh)]])
        yval = np.concatenate([Ay[tmp1][athresh:(2*athresh)],By[tmp2][bthresh:(2*bthresh)]])
        print "Split the original set of %d into %d and %d"%(X.shape[0],Xtrain.shape[0],Xval.shape[0])
        return Xtrain, ytrain, Xval, yval
    
    # Tracking progress using a dictionary
    large_dictionary = {}
    for i,gamma in enumerate(gammas):
        for j,C in enumerate(Cs):
            large_dictionary[(i+1)*1000+(j+1)]=[]

    
    our_seed=seed
    np.random.seed(our_seed)
    start_time=time.time()
    
    for k in range(num_runs):
        
        # We need balanced samples for this to make sense
        print "Starting on run %d out of %d. Current running time is %d seconds"%(k+1,num_runs,time.time()-start_time)
        # So first we need the data
        X_train, y_train, X_train_cv, y_train_cv = split(X_train_big, y_train_big, training_size)
        X_train = np.float64(X_train) # Convert everything to float 64, because regular float causes problems for some reason
        means = np.mean(X_train,axis=0)
        devs = np.std(X_train,axis=0)+1e-12 # Prevent divide by 0
        X_train = X_train - means
        X_train = X_train/devs
        
        X_train_cv = (np.float64(X_train_cv) - means )/(1e-12+devs)
        #print X_train_small.shape
        ROIs = []
        for i,gamma in enumerate(gammas):
            start = time.time()
            for j,C in enumerate(Cs):
                key = (i+1)*1000+(j+1)
                classifier = svm.SVC(kernel = 'rbf',gamma =gamma, C=C, class_weight = {0:1, 1:3.33})
                classifier.fit(X_train,y_train)
                # Let's see what the decision function is like
                df = classifier.decision_function(X_train_cv) # The decision function
                # So let's find the best threshold for marking something as positive
                best_t=-1
                best_ROI = -1e12
                our_costs=[]
                for t in thresholds:
                    y_hat = df>t
                    cost = dummy.bads_costs(y_train_cv,y_hat)
                    if(cost>best_ROI):
                        best_ROI=cost
                        best_t=t
                    our_costs.append(cost)
                y_hat = df>0.
                cost = dummy.bads_costs(y_train_cv,y_hat)
                if(print_conf):
                    print confusion_matrix(y_train_cv,y_hat)
                    print "ROI = %.4f with identified %d of %d as ret for g = %.4f  and C = %.1f"%(cost,np.sum(y_hat>0), np.sum(y_train_cv>0),np.log10(gamma),np.log10(C))
                
                if(show_T):
                    plt.figure(figsize=(7,7))
                    plt.plot(thresholds,our_costs)
                    plt.scatter(best_t,best_ROI,s=20)
                    plt.ylim([0.,1.5])
                    plt.ylabel("ROI")
                    plt.xlabel("Score Threshold")
                    plt.title("Optimum of %.4f with g=%.5f , c=%.2f and t= %.3f"%(best_ROI,gamma, C,best_t))
                    plt.show()
                large_dictionary[key].append(cost) # This is the cost associated with t=0, not the last t evaluated
        
    # Now let's see what the best combinations are
    # We'll fill a matrix with the combinations
    scores = np.zeros((gammas.shape[0],Cs.shape[0]))
    for key in large_dictionary.keys():
        i = int(key)/1000 - 1
        j = int(key)%1000 - 1
        tmp = large_dictionary[key]
        scores[i,j] = sum(tmp)/len(tmp) # Get the average

    #plt.figure(figsize=(20,20))
    #plt.imshow(scores,cmap='viridis')
    #plt.colorbar()
    
    # This was confusing

    #This gives us the overview
    #Let's also include a scatter plot of the combinations with the colors present
    gammas_flattened = []
    Cs_flattened = []
    intensity = []
    for i,gamma in enumerate(gammas):
        for j,C in enumerate(Cs):
            gammas_flattened.append(np.log10(gamma))
            Cs_flattened.append(C)
            intensity.append(scores[i,j])
    plt.figure(figsize=(7,7))
    plt.scatter(gammas_flattened,np.log10(Cs_flattened),c=intensity,cmap='viridis',s=60)
    plt.title("Average ROI for given C / gamma combination")
    plt.xlabel("log_10(gamma)")
    plt.ylabel("log_10(C)")
    plt.colorbar()
    return gammas_flattened, np.log10(Cs_flattened),intensity
```

```python
gammas_flattened, Cs_flattened,aaaintensity=evaluate_SVMs(training_size=1000,gammas=np.logspace(-5,-3,5),Cs=np.logspace(-1,2.5))
```

```python
gammas_flattened, Cs_flattened,aaaintensity=evaluate_SVMs(training_size=10000,gammas=[1e-4],Cs=[50])
```
