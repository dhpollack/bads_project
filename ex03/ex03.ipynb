```r
# 3.1
# 1)
df = read.csv("../ex01/Loan_Data.csv", sep=";")
df$YOB = as.factor(df$YOB)
df$PHON = as.factor(df$PHON)
df$BAD = as.factor(df$BAD)
loans = df
# 2, 3, 4, 5)
ggplot(data=loans, aes(x = dINC_A, y=dINC_SP)) + geom_point() + geom_smooth(method=lm)
```

```r
# 3.2
# 1)
summary(loans)
# 2)
ggplot(data=loans, aes(dINC_A)) + geom_histogram(aes(y=..density..)) + geom_density()
# 3)
boxplot(loans$dINC_A)
# 4)
dINC_A_sum = table(summary(loans$dINC_A))
#attributes(dINC_A_sum)
dINC_A_IQR = IQR(loans$dINC_A)
loans$dINC_A[which(loans$dINC_A > mean(loans$dINC_A) + 1.5*dINC_A_IQR)] = as.numeric(names(dINC_A_sum)[5])
summary(loans)
```

```r
# 3.3
loans_vt = sapply(loans, class)

kmodel = kmeans(loans[names(which(loans_vt == "numeric"))], 5, iter.max=50)
clusters = kmodel$cluster

l = list()
withins = c()
for(i in 1:7) {
    l[[i]] = kmeans(loans[names(which(loans_vt == "numeric"))], i, iter.max=50)
    withins = c(withins, l[i][[1]]$tot.withinss)
}

ggplot(data=data.frame(x=c(1:7), y=withins / sum(withins)),aes(x=x, y = y)) + geom_line() + ylim(0, 1.05*max(withins / sum(withins)))

# use 2 or 4 clusters
```
