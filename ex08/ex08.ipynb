````r
library("caret")
library("randomForest")
library("foreach")
````

````r
source('../helpfuncs/helpfuncs.R')
loans = get.loan.dataset('../data/Loan_Data.csv')
dim(loans)
````

````r
idx.train_valid_ = createDataPartition(y = loans$BAD, p = 0.8, list = F)
train_valid_ = loans[idx.train_valid_,]
test_   = loans[-idx.train_valid_, ]

model.control = trainControl(method="cv", number = 5, classProbs = T, summaryFunction = twoClassSummary, 
                             returnData = F)
rf.params = expand.grid(mtry = seq(1,5))
model.rf = train(BAD~., data = train_valid_, method = "rf", trace = T, 
                 tuneGrid = rf.params, trControl = model.control, 
                 metric = "ROC")
table(predict(model.rf, newdata=test_))
table(test_$BAD)
````

````r
print(model.rf)
````

````r
library("pROC")
yhat.rf = predict(model.rf, newdata = test_, type = "prob")[,1]
e.rf = roc(test_$BAD, yhat.rf)
auc(e.rf)
plot.roc(e.rf)
hist(yhat.rf)
````

````r
train_valid_.shf = train_valid_[sample(nrow(train_valid_)),]
k = 5
folds = cut(1:nrow(train_valid_.shf), breaks=k, labels = F)
#mtrySpan = seq(1,10)
#ntreeSpan = seq(100,1000,100)
mtrySpan = seq(1,6)
ntreeSpan = seq(100,600,100)

params.grid.rf = expand.grid(mtry = mtrySpan, ntree = ntreeSpan)
errors.rf = array(rep(NaN, k*length(mtrySpan)*length(ntreeSpan)),c(k,length(mtrySpan),length(ntreeSpan)))
t = foreach(i = 1:k) %do% {
    fold = which(folds == i)
    train_ = train_valid_.shf[-fold,]
    valid_ = train_valid_.shf[fold,]
    foreach(j = 1:nrow(params.grid.rf)) %do% {
        param = params.grid.rf[j,]
        model.rf = randomForest(BAD~., data = train_, 
                                summaryFunction = twoClassSummary, type = "regression",
                                mtry = param$mtry, ntree = param$ntree)
        yhat = predict(model.rf, newdata=valid_, type="prob")[,1]
        auc(valid_$BAD, yhat)
    }
}
````

````r
loans.pca = get.loan.dataset.nofact(fpath='../data/Loan_Data.csv')
loans.pca = loans.pca[-idx.train_valid_, ]
numcols = sapply(loans.pca,is.numeric)
loans_scaled = scale(loans.pca[,numcols])

model.pca = prcomp(loans_scaled[,-(ncol(loans_scaled)-1)]) # removing BAD column
````

````r
train_valid_2 = train_valid_
train_valid_2$BAD = as.numeric(train_valid_$BAD) - 1
test_2 = test_

model.svm = svm(BAD~., data=train_valid_2, cost=10^0, gamma = 0.25, type="eps-regression", probability = TRUE)
yhat.svm = predict(model.svm, newdata = test_2, probability = TRUE, decision.values = FALSE)
e.svm = roc(as.numeric(test_$BAD) - 1, yhat.svm)
auc(e.svm)
plot.roc(e.svm)
hist(yhat.svm)
````
